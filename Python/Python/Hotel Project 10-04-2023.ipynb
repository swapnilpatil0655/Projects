```python
# Hotel 
flag = True
def Welcome():
    print("Welcome to the Taj ")
    Menu()
      
def Menu ():
    global flag
    while flag:
        print("press 1. for Veg")
        print("press 2. for Non-veg")
        print("press 3. for Bill")
        print("press4. for exit")
        opt=int(input())
        dish(opt)

def dish(opt):
    global flag
    if(opt==1):
        print("Veg Dish name:")
        print("1. for Mix-veg")
        print("2. for Dal")
        print("3. for Rice")
        veg_opt=int(input())
        veg(veg_opt)
        
    elif(opt==2):
        print("Non-veg Dish name :")
        print("1. for Chicken")
        print("2. for Biryani")
        print("3. for Shawarma ")
        non_veg_opt=int(input())
        non_veg(non_veg_opt)
        
    elif (opt==3):
        final_Bill()
    
    elif (opt==4):
        print("Thank you for visting")
        final_Bill()
        flag=False
    
        
        
def veg(veg_opt):
    if(veg_opt==1):
        print("Mix-veg ........250 ")
        qty=int(input("Enter the total Quantity : "))
        price=int(input("Enter the Price :"))
        print("Mix_veg","qty=",qty,"Price=",price,"=",qty*price)
        Update_Bill(qty,price)
            
    elif(veg_opt==2):
        print("Dal ..........100 ")
        qty=int(input("Enter the total Quantity : "))
        price=int(input("Enter the Price :"))
        print("Dal","qty=",qty,"Price=",price,"=",qty*price)
        Update_Bill(qty,price)
        
    elif(veg_opt==2):
        print("Rice ..........150 ")
        qty=int(input("Enter the total Quantity : "))
        price=int(input("Enter the Price :"))
        print("Rice","qty=",qty,"Price=",price,"=",qty*price)
        Update_Bill(qty,price)
        
def non_veg(non_veg_opt):
    if(non_veg_opt==1):
        print("Chicken......250")
        qty=int(input("Enter the total Quantity : "))
        price=int(input("Enter the Price :"))
        print("Chicken","qty=",qty,"Price= ",price,"=",qty*price)
        Update_Bill(qty,price)
    
    elif(non_veg_opt==2):
        print("Biryani......350")
        qty=int(input("Enter the total Quantity : "))
        price=int(input("Enter the Price :"))
        print("Chicken","qty=",qty,"Price= ",price,"=",qty*price)
        Update_Bill(qty,price)
        
    elif(non_veg_opt==2):
        print("shawarma......200")
        qty=int(input("Enter the total Quantity : "))
        price=int(input("Enter the Price :"))
        print("Chicken","qty=",qty,"Price= ",price,"=",qty*price)
        Update_Bill(qty,price)
        
        
Bill=0
def Update_Bill(qty,price):
    global Bill
    Bill=Bill+(qty*price)
    final_Bill()
    
def final_Bill():
    print("Final Bill",Bill)
    
def exit():
    print("thanks for visiting")
    flag=False

Welcome()

```
