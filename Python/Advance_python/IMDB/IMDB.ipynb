# Q.4 Make a table of all the producers and drectors of each movie. Find the top 3 producers whohave produced with highest average ROI ?

```python
crew_df = df[~df['crew'].isna()]

def prod_lst(index):
  movie = crew_df.iloc[index]
  crew_lst = movie.loc['crew']
  producer_lst = []
  for elem in crew_lst:
    if elem['job'] == 'Producer':
      producer_lst.append(elem['name'])
  return producer_lst
```

```python
prod_lst(20)
```

```python
def direc_lst(index):
  movie = crew_df.iloc[index]
  crew_lst = movie.loc['crew']
  director_lst = []
  for elem in crew_lst:
    if elem['job'] == 'Producer':
      director_lst.append(elem['name'])
  return director_lst
```

```python
direc_lst(20)
```

```python
table = pd.DataFrame(columns = ['Movie Title','Producers','Director','ROI'])
```

```python
for index,row in crew_df.iterrows():
  try:
    table = table.append({'Movie Title':crew_df.loc[index,'original_title'],
                         'Producers':prod_lst(index),
                          'Director' :direc_lst(index),
                          'ROI':crew_df.loc[index,'roi']},
                         ignore_index = True)
  except:
    continue
```
